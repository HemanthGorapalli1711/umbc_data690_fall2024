```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# Define ASEAN countries
asean_countries = ['Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam']

# Choose a different year
year = 2018

# Filter the dataframe
df_filtered = df[(df['Country'].isin(asean_countries)) & (df['Year'] == year)][
    ['Country', 'GDP per capita (current US$)', 'Life expectancy at birth, total (years)', 'Population, total']
].dropna()

# Create a bubble chart with Plotly Express
fig = px.scatter(df_filtered,
                 x='GDP per capita (current US$)',
                 y='Life expectancy at birth, total (years)',
                 size='Population, total',
                 color='Country',
                 title=f'GDP per capita vs Life expectancy ({year}) for ASEAN Countries',
                 labels={'GDP per capita (current US$)': 'GDP per Capita (US$)',
                         'Life expectancy at birth, total (years)': 'Life Expectancy (Years)'},
                 hover_name='Country',
                 size_max=50,
                 log_x=True)  # Use log scale for GDP per capita

fig.update_layout(legend_title_text='Country')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# List of the seven regions
regions = ['East Asia & Pacific', 'Europe & Central Asia', 'Latin America & Caribbean',
           'Middle East & North Africa', 'North America', 'South Asia', 'Sub-Saharan Africa']

# Aggregate data
df_aggregated = df[df['Region'].isin(regions)].groupby(['Year', 'Region']).agg(
    avg_gdp_per_capita=('GDP per capita (current US$)', 'mean'),
    avg_life_expectancy=('Life expectancy at birth, total (years)', 'mean'),
    total_population=('Population, total', 'sum')
).reset_index()

# 1. Line chart showing the trend of Life Expectancy over time for all regions
fig1 = px.line(df_aggregated,
               x='Year',
               y='avg_life_expectancy',
               color='Region',
               title='Trend of Average Life Expectancy Over Time for Seven World Regions',
               labels={'avg_life_expectancy': 'Average Life Expectancy (Years)', 'Year': 'Year'},
               line_dash='Region',
               markers=True)

fig1.update_layout(legend_title_text='Region')
fig1.show()

# 2. Bubble chart comparing GDP per capita and life expectancy for a specific year
year = 2020  # You can change this to any year in the dataset

# Filter data for the selected year
df_year = df_aggregated[df_aggregated['Year'] == year]

fig2 = px.scatter(df_year,
                  x='avg_gdp_per_capita',
                  y='avg_life_expectancy',
                  size='total_population',
                  color='Region',
                  hover_name='Region',
                  title=f'Average GDP per Capita vs Average Life Expectancy ({year}) for Seven World Regions',
                  labels={'avg_gdp_per_capita': 'Average GDP per Capita (current US$)',
                          'avg_life_expectancy': 'Average Life Expectancy (Years)',
                          'total_population': 'Total Population'},
                  size_max=60,
                  log_x=True)

fig2.update_layout(legend_title_text='Region')
fig2.show()

# Print the aggregated data for verification
print(df_year)
```
