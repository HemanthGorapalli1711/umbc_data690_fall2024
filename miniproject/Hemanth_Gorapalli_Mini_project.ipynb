## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

This code categorizes countries into three groups: G7, BRICS, or Rest of the World, based on their country codes.

1. **G7_COUNTRIES** and **BRICS_COUNTRIES** are lists of 3-letter country codes for G7 and BRICS nations.
2. The **assign_group** function assigns a group based on the country code in each row.
3. It checks if the country code exists in the G7 or BRICS lists and returns the corresponding group.
4. If the country code is not in either list, it returns "REST OF WORLD."
5. The **apply()** function is used to apply this categorization to every row of the `data` DataFrame.
6. The function is applied across the rows (`axis=1`), creating a new "Group" column.
7. The **data[['Country Name', 'Country Code', 'Group']]** command selects the relevant columns.
8. The new "Group" column shows each country's assigned group.
9. The output displays the dataset with the country names, codes, and their groups.
10. This approach helps classify countries based on predefined lists.

```python
# Define G7 and BRICS countries using their 3-letter codes
G7_COUNTRIES = ["USA", "CAN", "GBR", "FRA", "DEU", "ITA", "JPN"]
BRICS_COUNTRIES = ["BRA", "RUS", "IND", "CHN", "ZAF"]
```

```python
# Function to categorize countries into G7, BRICS, or Rest of World
def assign_group(row):
    country_code = row['Country Code'].strip().upper()
    if country_code in G7_COUNTRIES:
        return "G7"
    elif country_code in BRICS_COUNTRIES:
        return "BRICS"
    else:
        return "REST OF WORLD"

# Apply the function to create the "Group" column
df['Group'] = df.apply(assign_group, axis=1)

# Display the entire dataset with the new 'Group' column
print("Complete Dataset with Categorization:")
df[['Country Name', 'Country Code', 'Group']]
```

## Step 5: Aggregate Data

	Grouping by 'Group' and 'Year': We group the data based on the Group and Year columns.

	Aggregation Functions:
  
  For GDP: We use sum() to aggregate the GDP values (from the column GDP ($Billions)).

  For GDP Per Capita: We use mean() to calculate the average GDP per capita (from the column GDP Per Capita ($)).

	Resetting Index: The reset_index() method flattens the DataFrame after aggregation so that "Group" and "Year" remain as columns.

	This will produce an aggregated dataset with the total GDP and the average GDP per capita, grouped by "Group" and "Year".


```python

# Aggregate GDP (sum) and GDP Per Capita (mean) at the Group level for each year
aggregated_data = df.groupby(['Year', 'Group']).agg({
    'GDP ($Billions)': 'sum',
    'GDP Per Capita ($)': 'mean'
}).reset_index()
aggregated_data.rename(columns={'GDP ($Billions)': 'Total GDP', 'GDP Per Capita ($)': 'Average GDP Per Capita'}, inplace=True)

# Display the aggregated data
aggregated_data
```

## Step 6 - Calculate Percentage Share of GDP

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

## Step 6: Calculate Percentage Share of GDP

We begin by aggregating the data at the "Group" and "Year" levels, summing GDP and calculating the average GDP per capita for each group within each year. Next, we calculate the total GDP for each year across all groups, which is used to determine each group's percentage share of the total GDP. This percentage share is computed by dividing each group's GDP by the total GDP for that year and multiplying by 100.

Afterward, the data is further aggregated by "Group" to calculate the total GDP, average GDP per capita, and the average percentage share of GDP for each group across all years. We display two separate outputs: one showing the aggregated data by group and year, and another showing the aggregated summary for each group over the entire time period. This provides both a detailed yearly breakdown and an overall group-level summary.


```python
# Step 1: Aggregate GDP and GDP Per Capita by Group and Year
aggregated_data = df.groupby(['Group', 'Year']).agg(
    GDP_sum=('GDP ($Billions)', 'sum'),
    GDP_per_capita_mean=('GDP Per Capita ($)', 'mean')
).reset_index()

# Step 2: Calculate the total GDP for each year
total_gdp_per_year = aggregated_data.groupby('Year')['GDP_sum'].transform('sum')

# Step 3: Calculate the percentage share of GDP for each group per year
aggregated_data['GDP_percentage_share'] = (aggregated_data['GDP_sum'] / total_gdp_per_year) * 100

# Display the aggregated data with GDP percentage share
print("Aggregated Data by Group and Year:")
aggregated_data
```

```python
# Step 4: Aggregate the data further by Group to get overall data for each group
group_aggregated_data = aggregated_data.groupby('Group').agg(
    Total_GDP=('GDP_sum', 'sum'),
    Avg_GDP_per_capita=('GDP_per_capita_mean', 'mean'),
    Avg_GDP_percentage_share=('GDP_percentage_share', 'mean')
).reset_index()

# Separate output for each group
print("\nAggregated Data for Each Group:")
group_aggregated_data
```

## Step 7 - Plot the GDP Percentage Share Over Time

It is an interactive line plot using Plotly to visualize the **GDP percentage share over time** for different groups. It iterates through each group, adding a line trace with markers for each group's data.
```python
import plotly.graph_objects as go

# Create the line plot for GDP percentage share over time for each group
fig = go.Figure()

# Add a trace for each group to the plot
for group in aggregated_data['Group'].unique():
    group_data = aggregated_data[aggregated_data['Group'] == group]
    fig.add_trace(go.Scatter(
        x=group_data['Year'],
        y=group_data['GDP_percentage_share'],
        mode='lines+markers',  # Show both lines and markers for data points
        name=group  # Use the group name as the legend label
    ))

# Update the layout of the plot for better visualization with a gradient background
fig.update_layout(
    title='GDP Percentage Share Over Time by Group',
    xaxis_title='Year',
    yaxis_title='GDP Percentage Share (%)',
    plot_bgcolor='rgba(135,206,235,0.7)',  # Skyblue with some transparency for plot area
    paper_bgcolor='rgba(255,255,255,0.8)',  # White with some transparency for paper background
    title_font=dict(size=20, color='black'),
    xaxis=dict(tickfont=dict(color='black')),
    yaxis=dict(tickfont=dict(color='black')),
    legend_title='Group',
    hovermode='x unified',  # Improve hover behavior for better comparison
)

# Show the plot
fig.show()
```
